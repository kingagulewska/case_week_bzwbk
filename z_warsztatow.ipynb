```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
```

```python
dane = pd.read_csv("train.csv")
```

```python
dane = dane.replace(-999999,2)
dane.describe()
```

```python
cnt_uniq = dane.apply(lambda x: len(x.unique()))
uniq = cnt_uniq[cnt_uniq == 1].index.tolist()
len(uniq)
```

```python
dane = dane.drop(uniq, axis=1)
dane.head()
```

```python
from itertools import combinations

col_por = list(combinations(dane.columns.tolist(),2))
same_col = []

for col in col_por:
    is_taka_sama = np.array_equal(dane[col[0]], dane[col[1]])
    if is_taka_sama:
        same_col.append(list(col))
len(same_col)
    
```

```python
to_drop = [item[0] for item in same_col]
dane = dane.drop(to_drop, axis=1)

dane.head()
```

```python
# tylko top 10 cech
dane.shape
```

```python
from sklearn.utils import shuffle
dane = shuffle(dane)

zero_only = dane.loc[dane['TARGET'] == 0]
jeden_only = dane.loc[dane['TARGET'] == 1]

len(zero_only), len(jeden_only)
```

```python
train = pd.concat([jeden_only.iloc[:2000], zero_only.iloc[:2000]])
test = pd.concat([jeden_only.iloc[2000:], zero_only.iloc[2000:3000]])

train.shape, test.shape
```

```python
train = shuffle(train)
test = shuffle(test)

y_train = train['TARGET']
y_test = test['TARGET']

X_train = train.drop(['ID','TARGET'], axis = 1)
X_test = test.drop(['ID', 'TARGET'], axis = 1)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

from sklearn.metrics import accuracy_score, precision_score, confusion_matrix
from sklearn.model_selection import cross_val_score

classifiers = [KNeighborsClassifier(11),
       LogisticRegression(),
              DecisionTreeClassifier(),
              RandomForestClassifier(),
              GradientBoostingClassifier()]

name = []
result_cv = []

for clf in classifiers:
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)

    print(clf.__class__.__name__)
    name.append(clf.__class__.__name__)
    
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    res_cv = cross_val_score(clf, X_train, y_train, cv=10, scoring="precision")
    result_cv.append(res_cv)
    
    print('Accuracy : {:.2%}'.format(acc))
    print('Precision : {:.2%}'.format(prec))
    print('Precision z cross validacji : {:.2%}'.format(res_cv.mean()))
    print()
    
```

```python
fig = plt.figure(figsize=(14,8))
ax = fig.add_subplot(111)
plt.boxplot(result_cv)
ax.set_xticklabels(name)
plt.show()
```
